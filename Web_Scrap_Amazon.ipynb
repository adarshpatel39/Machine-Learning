# Task
Extract all links from the HTML content in the file "example.html" and save them to an Excel file named "extracted_links.xlsx".

## Extract links

### Subtask:
Extract all the links (href attributes) from the HTML.


**Reasoning**:
The HTML content is already loaded in the `soup` object from the previous steps. I need to find all 'a' tags and extract their 'href' attributes, storing them in a list.



```python
links = []
for link in soup.find_all('a'):
    href = link.get('href')
    if href:
        links.append(href)
```

## Create dataframe

### Subtask:
Create a pandas DataFrame from the extracted links.


**Reasoning**:
Create a pandas DataFrame from the extracted links.



```python
import pandas as pd

df_links = pd.DataFrame(links, columns=['href'])
```

```python
df_links
```

## Export to excel

### Subtask:
Save the DataFrame to an Excel file.


**Reasoning**:
Save the `df_links` DataFrame to an Excel file without the index.



```python
df_links.to_excel("extracted_links.xlsx", index=True)
```
