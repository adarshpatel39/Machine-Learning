```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/titanic.csv')
df.sample(5)
```

**1**. **Categorical** **Data**

     a.Countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Pclass', data=df)
plt.title('Count of Survived Passengers')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%2f')
x_label=['Survived','Not Survived']
plt.show()
```

**Numerical Data**

```python
import seaborn as sns
sns.kdeplot(data=df, x='Age')
plt.title('Probability Density Function of Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
import seaborn as sns
sns.displot(data=df, x='Age', kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
